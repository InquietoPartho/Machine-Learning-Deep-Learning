Handle Imbalanced Dataset

A dataset with an unequal distribution

```python
import pandas as pd
import numpy as np
```

```python
credit_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_data.head()
```

```python
credit_data.tail()
```

```python
credit_data.shape
```

```python
#distribution of the two classes

credit_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 ---> Legit Transaction


1 ---> Fraudulent Transaction

```python
#separating the legit and fraudulent transactions

legit = credit_data[credit_data.Class == 0.0]
fraud = credit_data[credit_data.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

#Under-sampling

Build a sample dataset containing similar distribution of legit and fraudulent Transactions

Number of Fraudulent Transaction --->178

```python
legit_sample = legit.sample(n=178)
```

```python
print(legit_sample)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
