```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles  = 'xy', scale=1,color = 'b')
plt.quiver(0,0,-3,4, scale_units = 'xy', angles  = 'xy', scale=1,color = 'r')
plt.quiver(0,0,-5,-6, scale_units = 'xy', angles = 'xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles  = 'xy', scale=1,color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles  = 'xy', scale=1,color = 'r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles  = 'xy', scale=1,color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles  = 'xy', scale=1,color = 'r')
plt.quiver(0,0,5,1, scale_units = 'xy', angles  = 'xy', scale=1,color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#Subtraction

```python
diff  = vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles  = 'xy', scale=1,color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles  = 'xy', scale=1,color = 'r')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles  = 'xy', scale=1,color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#DOT Product
a.b = |a||b|cosx

```python
a = np.array([2,3])
b = np.array([3,2])
c = np.dot(a,b)
print(c)
```

```python
a = np.array([3,2,5])
b = np.array([2,3,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

#Cross Product

a x b = |a||b|sinx

```python
cross = np.cross(a,b)
print(cross)
```
